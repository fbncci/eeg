```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#just use one csv as an example
filepath = "../../../groups/eeg/eeg-data/eeg1.csv"
```

```python
df = pd.read_csv(filepath)
```

```python
info_cols = ["time", "label_0", "label_1", "label_2", "Resp Effort-REF", "ECG EKG-REF"]
label_col = ["label_0", "label_1", "label_2"]
data_cols = list(set(df.columns).difference(info_cols))
```

```python
X = df[data_cols].values
y = df[label_col].values

max_val = []
labels = []

for row in X:
    max_val.append(max(row))

for row in y:
    if sum(row) == 3:
        labels.append(1)
    else:
        labels.append(0)

df['label'] = labels
df['max'] = max_val
```

```python
%matplotlib inline
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(15,15))

df['label'][:len(df.label_0.dropna())].plot(ax=axes[0])
df["max"][:len(df.label_0.dropna())].plot(ax=axes[1])
# df.label_0.plot(ax=axes[1])
# df.label_1.plot(ax=axes[2])
# df.label_2.plot(ax=axes[3])
# df["ECG EKG-REF"][:len(df.label_0.dropna())].plot(ax=axes[5])
# df["ECG EKG-REF"][:len(df.label_0.dropna())].plot(ax=axes[6])

fig.legend()
```
